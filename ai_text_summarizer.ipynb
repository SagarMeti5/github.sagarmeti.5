```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="t5-small", tokenizer="t5-small")
```

```python
text = """
Accenture is a global professional services company with leading capabilities in digital, cloud and security.
Combining unmatched experience and specialized skills across more than 40 industries,
it offers Strategy and Consulting, Interactive, Technology and Operations services.
Accenture works at the intersection of business and technology to help clients improve performance.
"""

summary = summarizer(text, max_length=60, min_length=20, do_sample=False)
print("Summary:", summary[0]['summary_text'])
```

```python
def generate_summary(input_text, max_len=60, min_len=20):
    summary = summarizer(input_text, max_length=max_len, min_length=min_len, do_sample=False)
    return summary[0]['summary_text']

# Test it
print(generate_summary("Artificial intelligence is transforming industries by automating tasks, improving decision making, and enabling new business models."))
```

```python
user_text = input("Paste your text to summarize:\n")
print("\nSummary:\n", generate_summary(user_text))
```
